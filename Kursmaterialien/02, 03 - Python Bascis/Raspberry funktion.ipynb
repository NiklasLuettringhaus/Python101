```python
#!/usr/bin/python
import RPi.GPIO as GPIO
from tkinter import *
import time

GPIO.setmode(GPIO.BCM)
LED = [26,13,6,5]
for i in LED:
    GPIO.setup(i, GPIO.OUT, initial=False)

w=5
t=0.2
isrunning = 1

muster =[
    ("Lauflicht links", 1),
    ("Blinken", 2),
    ("Lauflicht recht",3)
]
root = Tk()
root.title("Lauflicht")
v = IntVar()
v.set(1)

def Aus():
    trigger = 0
    trigger = trigger + -1

def LedEin():
    if isrunning == 0:
        return()
    else:
        e = v.get()
        if e == 1:
            for i in range(w):
                for j in range(4):
                    GPIO.output(LED[j], True)
                    time.sleep(t)
                    GPIO.output(LED[j], False)
        elif e == 2:
            for i in range(w):
                for j in range(4):
                    GPIO.output(LED[j], True)
                time.sleep(t)
                for j in range (4):
                    GPIO.output(LED[j], False)
                time.sleep(t)
        else:
            for i in range(w):
                for j in range(4):
                    GPIO.output(LED[3-j], True)
                    time.sleep(t)
                    GPIO.output(LED[3-j], False)


Label(root,
        text="button klicken um Lauflicht zu starten").pack()

for txt, m in muster:
    Radiobutton(root,
                text=txt,
                variable = v,
                value = m).pack(anchor=W)

Button(root,
        text="Start",
        command=LedEin).pack(side=LEFT)
Button(root,
        text="Abbruch",
        command=Aus).pack(side=RIGHT)

root.mainloop()

GPIO.cleanup()
```
